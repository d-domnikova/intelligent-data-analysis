```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
df = pd.read_csv('Students_Grading_Dataset.csv')
df.describe()
```

```python
df.describe(include=['object'])
```

```python
df.isna().sum()
```

```python
df_drop_indexes = df.dropna()

df_fin = df.copy()
df_fin.fillna(df.mean(numeric_only=True), inplace=True)
df_fin.fillna(df.mode().iloc[0], inplace=True)

df_fin.loc[len(df_fin)] = ['S6000','James','Williams','student5000@university.com','Male',
                               22,'Mathematics', 152.29, 7.03, 7.82, 84.22, 74.06,
                               223.99, 355.90, 56.09,'B', 82.2,'No','Yes','High School','Medium', 4, 1.7]
```

```python
df_fin.boxplot(by='Age', figsize=[18, 12], layout=[5,6])
```

```python
numeric = df_fin.select_dtypes('number').columns
iqr = df_fin.quantile(0.75, numeric_only=True) - df_fin.quantile(0.25, numeric_only=True)
lim = np.abs((df_fin - df_fin.median(numeric_only=True)) / iqr) < 2.22

df_fin.loc[:, numeric] = df_fin.where(lim, np.nan)
df_fin.dropna(inplace=True)
df_fin.boxplot(by='Age', figsize=[18, 12], layout=[5,6])
```

```python
normalized_values = MinMaxScaler().fit_transform(df_fin[numeric])
df_fin = pd.DataFrame(normalized_values, columns = numeric)
df_fin.describe()
```
